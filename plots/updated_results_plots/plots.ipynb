```python
####XGBOOST
score_time_report = pd.read_csv("xgboost_score_time_report.csv")

datasets = score_time_report["Dataset"].unique()
methods = score_time_report.groupby("Dataset", sort=False)["Method"].apply(list)
time = score_time_report.groupby("Dataset", sort=False)["Time"].apply(list)
score = score_time_report.groupby("Dataset", sort=False)["Score"].apply(list)

for i in range(0,4):
    score_ar = np.array(score[i])
    score_float = score_ar.astype(np.float64)
    
    time_ar = np.array(time[i])
    time_float = time_ar.astype(np.float64)
    
    scores = pd.DataFrame(dict(data=score_float))
    times = pd.DataFrame(dict(data=time_float))
#     times["data"] = (times.data / 60).round(2)
    
    fig, ax1 = plt.subplots(figsize=(8, 5))
    ax2 = ax1.twinx()
    ax1.set_ylabel("Accuracy", color="#0147AB", size=14)
    ax2.set_ylabel("Time", color="red", size=14)
    ax1.tick_params(axis="y", labelcolor="#0147AB", labelsize=12)
    ax2.tick_params(axis="y", labelcolor="red", labelsize=12)
    ax1.tick_params(axis="x", labelsize=12, pad=10)
#     ax1.set_ylim([0.1, 1])
#     ax2.set_ylim([0, 600])
    
    
    
    ax1.set_title(datasets[i] + " Xgboost", fontsize=14)

    ax1.bar(methods[0], scores["data"], width = 0.4,color=['orange', 'gray', 'olive', 'purple'])

    ax2.plot(methods[0], times["data"], color="red", marker="*")

plt.show()
```

```python
####CATBOOST
score_time_report = pd.read_csv("catboost_score_time_report.csv")

datasets = score_time_report["Dataset"].unique()
methods = score_time_report.groupby("Dataset", sort=False)["Method"].apply(list)

time = score_time_report.groupby("Dataset", sort=False)["Time"].apply(list)
score = score_time_report.groupby("Dataset", sort=False)["Score"].apply(list)

for i in range(0,4):
    score_ar = np.array(score[i])
    score_float = score_ar.astype(np.float64)
    
    time_ar = np.array(time[i])
    time_float = time_ar.astype(np.float64)
    
    scores = pd.DataFrame(dict(data=score_float))
    times = pd.DataFrame(dict(data=time_float))

    fig, ax1 = plt.subplots(figsize=(8, 5))
    ax2 = ax1.twinx()
    ax1.set_ylabel("Accuracy", color="#0147AB", size=14)
    ax2.set_ylabel("Time", color="red", size=14)
    ax1.tick_params(axis="y", labelcolor="#0147AB", labelsize=12)
    ax2.tick_params(axis="y", labelcolor="red", labelsize=12)
    ax1.tick_params(axis="x", labelsize=12, pad=10)
#     ax1.set_ylim([0.1, 1])
    
    ax1.set_title(datasets[i] + " Catboost", fontsize=14)

    ax1.bar(methods[i], scores["data"], width = 0.4,color=['orange', 'gray', 'olive', 'purple'])

    ax2.plot(methods[i], times["data"], color="red", marker="*")


plt.show()


```
